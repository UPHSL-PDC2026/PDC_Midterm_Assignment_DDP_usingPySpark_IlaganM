```python
from pyspark.sql.functions import col
from google.colab import drive

drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/PDC_CFILES/student_mental_health_burnout_1M.csv'


df = spark.read.csv(file_path, header=True, inferSchema=True)

df.show(5)
```

```python
filtered_df = df.filter(col("age") > 23)
filtered_df.show()
```

```python
grouped_df = df.groupBy("gender").count()
grouped_df.show()
```

```python
sorted_df = grouped_df.orderBy("count", ascending=False)
sorted_df.show()
```
